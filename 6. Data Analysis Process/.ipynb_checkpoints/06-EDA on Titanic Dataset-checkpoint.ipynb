## Feature Engineering

```python
# combine SibSp and Parch to make a new column 'Family_size'
df['family_size'] = df['SibSp'] + df['Parch']
```

```python
# Now we will engineer a new column by the name of family_type

def family_type(number):
    if number == 0:
        return 'Alone'
    elif number > 0 and number <=4:
        return 'Medium'
    else:
        return 'Large'
```

```python
df['family_type'] = df['family_size'].apply(family_type)
```

```python
df.head()
```

```python
df.drop(columns={'SibSp', 'Parch', 'family_size'}, inplace=True)
```

## Detecting Outliers

### Numeric Data
- If the data follows normal distribution, anything beyond `3SD - mean + 3SD` can be considerd as an `Outlier`
- If the data does not follow normal distribution, using boxplot we can eliminate points beyond `Q1 - 1.5` and `Q3 + 1.5 IQR`

### Categorical Data
- If the col is highly imbalanced for eg. male 100000 and female 2, then we can eliminate female colum from the dataset.

```python
# Handling outliers in age (Almost normal)

df[df['Age'] < (df['Age'].mean() + 3 * df['Age'].std())]
df.shape # before removing outliers
```

```python
df = df[df['Age'] < (df['Age'].mean() + 3 * df['Age'].std())]
df.shape # after removing outliers
```

```python
# Handling outliers from Fare column
# finding quartiles

Q1 = np.percentile(df['Fare'], 25)
Q3 = np.percentile(df['Fare'], 75)

outlier_low = Q1 - 1.5 * (Q3 - Q1)
outlier_high = Q3 + 1.5 * (Q3 - Q1)

df = df[(df['Fare'] > outlier_low) & (df['Fare'] < outlier_high)]
```
