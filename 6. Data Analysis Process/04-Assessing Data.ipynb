   You will definitely need to have a better of what the data is about

### Types of Unclean Data
   
   There are 2 kinds of Unclean data
    
   1. Dirty Data (Data with quality Issues):
       Dirty Data, also known as low quality data. Low quality data has `Quality Issues`
   2. Messy Data (Data with tidiness Issues):
       Messy Data, also know as Untidy data. Untidy data has `Structural Issues`
    

### 1. Dirty Data

When the data has following kinds of issues:

1.  Duplicate values
2. Missing values
3. Curropted data (Basicly data is not visible)
4. Inaccurate data

### 2. Messy Data

When the data has Structural issues it is know as `Messy Data` OR `Untidy Data`

Tidy Data has following properties:

1. Each variable forms a column
2. Each observation forms a row
3. Each observational unit forms a table

   - If the data voilates any one of the above it becames `Untidy Data`

### Types of Assessment

1. Manual Assessment- In this process you try to find Error is the dataset 'Manualy' by just lookin
2. Programmatic Assessment- In this process you use Pandas Functions such as info(), describe(), sample() etc...

# Quality Issues

### [Completeness]

### `Patients`:
- zip code is stored in Float datatype [Validity Issue]
- some of the zip codes is of 4 digits [Validity Issue]
- for [state] column sometimes full names are used and sometimes short forms/abbriviations are used.[Consistency Issue]
- [address, city, state, zip code, contact] all these columns has missing values [Completeness Issue]
- [assinged_sex], [birthdate] column has inaccurate datatype [Accuracy Issue]
- Erroneous data for weigts and heights [Accuracy Issue]
- Inconsistent way of storing phone numbers in [contact] column [Consistency Issue]
- Duplicate data and some place holder data in [address] column [Consistency Issue]
### `Treatments`:
- hba1c_change column has missing values [Completeness Issue]
- columns [novodra, auralin] has missing values also these columns have `u` attached to numbers. [Accuracy & Completeness Issue]
- Only 280 rows are there, instead of 350 (70 rows are missing). [Completeness Issue]
- Name and Surname column is in lowercase [Consistency Issue]
- columns [novodra, auralin] has inaccurate datatype [Validity Issues]
- In [hba1c_change] column some values are `9` instead of `4` [Accuracy Issue]
- In [auralin, novodra] column `dashes` are used instead of `NaN` [Consistency Issue]
### `Adverse Reactions`:
- Name and Surname column is in lowercase [Consistency Issues]
