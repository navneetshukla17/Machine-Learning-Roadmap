# Univariate Analysis

## 1. Categorical Data

### a. Countplot

```python
# countplot on Survived

sns.countplot(data=df, x='Survived')
```

```python
# countplot on Pclass

sns.countplot(data=df, x='Pclass')
```

```python
# countplot on Sex

sns.countplot(data=df, x='Sex')
```

```python
# countplot on Embarked

sns.countplot(data=df, x='Embarked')
```

### b. PieChart

```python
# piechart of Survived

df['Survived'].value_counts().plot(kind='pie', autopct='%.1f')
```

```python
# piechart of Pclass

df['Pclass'].value_counts().plot(kind='pie', autopct='%.1f')
```

```python
# Piechart of Sex

df['Sex'].value_counts().plot(kind='pie', autopct='%.1f')
```

```python
# Piechart on Embarked

df['Embarked'].value_counts().plot(kind='pie', autopct='%.1f')
```
